```python
import pathlib

import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data = pd.read_csv('dataset.csv', delimiter=',')
dataset = data.copy()
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
```

```python
dataset.isna().sum()
```

```python
dataset
```

```python
dataset.head()
```

```python
year = dataset.pop('Year_of_Start')
dataset["year2012"] = (year == 2012)*1.0
dataset["year2013"] = (year == 2013)*1.0
dataset["year2014"] = (year == 2014)*1.0
dataset["year2015"] = (year == 2015)*1.0
dataset["year2016"] = (year == 2016)*1.0

performance = dataset.pop('IndustryPerformance')
dataset["high_performance"] = (performance == "High")*1.0
dataset["medium_performance"] = (performance == "Medium")*1.0
dataset["low_performance"] = (performance == "Low")*1.0

performance = dataset.pop('CompetitiveLandscape')
dataset["high_comp"] = (performance == "High")*1.0
dataset["medium_comp"] = (performance == "Medium")*1.0
dataset["low_comp"] = (performance == "Low")*1.0
# sns.pairplot(train_dataset[["Year_of_Start","Empl_Cnt_Y1", "City", "Industry_Vertical","OperationalDimension","CustomerDimension","MarketDimension","ManamgementDimension","StrategicDimenision","FinanceDimension","HRDimension"]], diag_kind="kde")
```

```python
dataset.pop("Empl_Cnt_Y1")
dataset.pop("Invest_1st_Q_S")
dataset.pop("InvestmentType")
dataset.pop("Industry_Vertical")
dataset.pop("SubVertical")
dataset.pop("City")
dataset.pop("id")
```

```python
train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
test_dataset
```

```python
train_stats = train_dataset.describe()
# train_stats.pop("id")
train_stats.pop("status")
train_stats = train_stats.transpose()
```

```python
train_labels = train_dataset.pop("status")
test_labels = test_dataset.pop("status")
```

```python
def norm(x):
    return (x - train_stats["mean"]) / train_stats["std"]
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
# dataset
```

```python
normed_train_data
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation=tf.nn.relu, input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation=tf.nn.relu),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
```

```python
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch
  
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error')
  plt.plot(hist['epoch'], hist['mean_absolute_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_absolute_error'],
           label = 'Val Error')
  plt.legend()
  plt.ylim([0,1])
  
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error')
  plt.plot(hist['epoch'], hist['mean_squared_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_squared_error'],
           label = 'Val Error')
  plt.legend()
  plt.ylim([0,1])

plot_history(history)
```

```python
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

plot_history(history)
```

```python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=0)

print("Testing set Mean Abs Error: {:5.2f}".format(mae))
```

```python
test_predictions = model.predict(normed_test_data).flatten()

import numpy as np
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
```
